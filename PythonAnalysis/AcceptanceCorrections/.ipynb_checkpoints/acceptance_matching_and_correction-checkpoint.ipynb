```python


# ------------------------------------------------------------------------------- #
# Mar-24,2022
def plot_step_hist( x_arr, bins=[0.1,0.2,0.3], label="",
                    color='k', ScaleFactor=1, weights=None,
                                        do_plot_step=True,do_plot_errorbar=True,
                   density=False, capsize=2, capthick=2,
                   marker='o',linewidth=2, markersize=5,alpha=1.0):#{
    import numpy as np, matplotlib.pyplot as plt
    
    h,edges = np.histogram( x_arr, bins=bins, weights=weights )
    h_err = np.sqrt(h)

    if density==True:
        h_err = h_err/np.sum(h)
        h     = h / np.sum(h)

    h     = h*ScaleFactor
    h_err = h_err*ScaleFactor
    x     = (edges[1:] + edges[:-1])/2
    x_err = (edges[1:] - edges[:-1])/2
    if do_plot_step:
        plt.step ( x, h, color=color, where='mid', label=None if do_plot_errorbar==True else label, linewidth=linewidth )
    if do_plot_errorbar:
        plt.errorbar ( x=x, xerr=x_err, y=h, yerr=h_err, markersize=markersize,
                  color=color, marker=marker, linestyle='None',label=label,
                  capsize=capsize, capthick=capthick, linewidth=linewidth,alpha=alpha )
    return x,h,x_err,h_err
#}
# ------------------------------------------------------------------------------- #

```
