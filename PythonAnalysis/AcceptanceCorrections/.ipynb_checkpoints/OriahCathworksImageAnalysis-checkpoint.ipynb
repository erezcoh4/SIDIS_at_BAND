```python
from scipy.io import loadmat
import scipy as sp, numpy as np, matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
I1 = loadmat('/Users/erezcohen/Desktop/Projects/Oriah/Cathworks_Mar2022/II/I1.mat')
I2 = loadmat('/Users/erezcohen/Desktop/Projects/Oriah/Cathworks_Mar2022/II/I2.mat')
```

```python
I1_image = I1['I1']
I2_image = I2['I2']
```

```python
def draw_ROI_in_image( im, ax , do_add_monitor_plots=False ):
    '''
    return 
    ------
    [xy_bl,xy_br,xy_tl,xy_tr]  boundary points: bottom-left, bottom-right, top-left, top-right
    
    '''
    # print('draw_ROI_in_image( im )')    
    
    # flatten the image and create a distribution of the intensities to determine a frame threshold
    x = im.flatten();
    min_x = np.min(x)
    max_x = np.max(x)
    h,edges = np.histogram( x, bins = np.linspace(min_x,max_x,100) )    
    centers = (edges[1:]+edges[:-1])/2 # location of the bins in x (intensity bins)
    # find most prominent peaks
    peaks,_ = find_peaks( h , prominence=(1000,None))    
    
    # set the threshold between the first two peaks
    # the left-most one (the dark one) representing the dark part of the "optical frame"
    # and the right-most one (a brighter one) represents the next intensity "peak" in the ROI picture
    # average over peak locations
    threshold = (centers[peaks[0]]+centers[peaks[1]])/2    
    
    im_frame  = np.ma.masked_where(threshold>im, im)
    im_inside = np.ma.masked_where(threshold<im, im)
        
    # define the frame boundaries
    indices_frame_y,indices_frame_x = np.where( (0<im) & (im<threshold));
    h_indices_frame_x,idx_edges_x = np.histogram(indices_frame_x, bins=101)    
    Nx = len(h_indices_frame_x);
    h_indices_frame_y,idx_edges_y = np.histogram(indices_frame_y, bins=101)    
    Ny = len(h_indices_frame_y);
    
    # select left boundary as the left-most location of the frame pixels
    # start from the middle and go leftward
    left_boundary = 0
    for i in np.arange( int(Nx/2), 1, -1):
        #print(i,idx_edges_x[i],h_indices_frame_x[i],h_indices_frame_x[i-1])
        if float(h_indices_frame_x[i-1])/h_indices_frame_x[i] > 1.3:
            left_boundary = idx_edges_x[i]
            
    # select right boundary as the right-most location of the frame pixels
    # start from the middle and go rightwards
    right_boundary = left_boundary+1
    for i in np.arange( int(Nx/2), Nx-1, 1):
        if float(h_indices_frame_x[i+1])/h_indices_frame_x[i] > 1.3:
            right_boundary = idx_edges_x[i]

    # select upper boundary as the upper-most location of the frame pixels
    # start from the middle and go upward
    upper_boundary = 0
    for i in np.arange( int(Ny/2), 1, -1):
        if float(h_indices_frame_y[i-1])/h_indices_frame_y[i] > 1.3:
            upper_boundary = idx_edges_y[i]
            
    # select bottom boundary as the bottom-most location of the frame pixels
    # start from the middle and go downward
    bottom_boundary = upper_boundary+1
    for i in np.arange( int(Ny/2), Ny-1, 1):
        if float(h_indices_frame_y[i+1])/h_indices_frame_y[i] > 1.3:
            bottom_boundary = idx_edges_y[i]

    # return the four points that form the boundaries
    xy_bl = [left_boundary,bottom_boundary]
    xy_br = [right_boundary,bottom_boundary]
    xy_tl = [left_boundary,upper_boundary]
    xy_tr = [right_boundary,upper_boundary]
    boundary_pts = [xy_bl,xy_br,xy_tl,xy_tr]

            
    # plot boundaries on the image
    ax.plot( [left_boundary,left_boundary], [bottom_boundary,upper_boundary],'--w')
    ax.plot( [right_boundary,right_boundary], [bottom_boundary,upper_boundary] ,'--w')
    ax.plot( [left_boundary,right_boundary], [bottom_boundary,bottom_boundary] ,'--w')
    ax.plot( [left_boundary,right_boundary], [upper_boundary,upper_boundary] ,'--w')
    ax.scatter( left_boundary,  bottom_boundary , marker='o' , facecolor='y', edgecolor='k')
    ax.scatter( right_boundary, bottom_boundary , marker='o' , facecolor='y', edgecolor='k')
    ax.scatter( left_boundary,  upper_boundary , marker='o' , facecolor='y', edgecolor='k')
    ax.scatter( right_boundary, upper_boundary , marker='o' , facecolor='y', edgecolor='k')
           
        
    
    if do_add_monitor_plots:
    
        fig2 = plt.figure(figsize=(8,5))
        ax = fig2.add_subplot(2,2,1);
        ax.step( centers, h,where='mid')
        plt.plot( centers[peaks], h[peaks], "x", markersize=10)
        plt.plot( [threshold,threshold], ax.get_ylim(), '--r')
        ax.set_title('Intensity distribution')
        ax.set_xlabel('Intensity')
        ax.set_ylabel('Counts')

        ax = fig2.add_subplot(2,2,2);
        plt.imshow( im_inside );
        plt.plot( [left_boundary,left_boundary], [bottom_boundary,upper_boundary],'--w')
        plt.plot( [right_boundary,right_boundary], [bottom_boundary,upper_boundary] ,'--w')
        plt.plot( [left_boundary,right_boundary], [bottom_boundary,bottom_boundary] ,'--w')
        plt.plot( [left_boundary,right_boundary], [upper_boundary,upper_boundary] ,'--w')


        ax = fig2.add_subplot(2,2,3);
        plt.hist( indices_frame_x, bins=101 );
        plt.plot( [left_boundary,left_boundary], ax.get_ylim(),'--r')
        plt.plot( [right_boundary,right_boundary], ax.get_ylim(),'--r')
        ax.set_xlabel('pixel x location')
        ax.set_ylabel('Counts')


        ax = fig2.add_subplot(2,2,4);
        plt.hist( indices_frame_y, bins=101 );
        plt.plot( [upper_boundary,upper_boundary], ax.get_ylim(),'--r')
        plt.plot( [bottom_boundary,bottom_boundary], ax.get_ylim(),'--r')
        ax.set_xlabel('pixel y location')
        ax.set_ylabel('Counts')

        plt.tight_layout()    
    
    return boundary_pts
```

```python
fig = plt.figure(figsize=(20,12))
for frame_idx in range(24):
    ax = fig.add_subplot(4,6,frame_idx+1)
    im = I1_image[:,:, frame_idx]
    plt.imshow( im );
    draw_ROI_in_image( im, ax, False );
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(20,12))
for frame_idx in range(24): # 
    ax = fig.add_subplot(4,6,frame_idx+1)
    im = I2_image[:,:, frame_idx]
    plt.imshow( im );
    draw_ROI_in_image( im, ax, do_add_monitor_plots=False );
plt.tight_layout()
```
