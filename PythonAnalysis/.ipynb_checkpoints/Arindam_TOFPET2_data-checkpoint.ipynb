```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

```python
data_path = '/Users/erezcohen/Desktop/data/PETsys/BoxSi_proto4.0/Arindam_data/'
data,times,qdc,N = dict(),dict(),dict(),dict()
for ch in [640,764]:
    data[ch] = pd.read_csv(data_path + 'test_%d.txt'%ch,delimiter='\t')
    times[ch] = np.array(data[ch].time)
    qdc[ch]   = np.array(data[ch].qdc)
    N[ch]     = len(data[ch])
print('Done loading data')
```

```python
dt_coincidenece = 100e3; # 100 ns
ch_1 = 640
ch_2 = 764
qdc_coincidence_1, qdc_coincidence_2 = [],[]

for i_1 in range(N[ch_1]):
    for i_2 in range(N[ch_2]):
        if np.abs(times[ch_1][i_1] - times[ch_2][i_2]) < dt_coincidenece:
            qdc_coincidence_1.append(qdc[ch_1][i_1])
            qdc_coincidence_2.append(qdc[ch_2][i_2])
            
print('Done collecting %d coincidence events within dt < %d ns'%(len(qdc_coincidence_1),dt_coincidenece/1e3))
```

```python
fig,ax = plt.subplots()
plt.scatter( qdc_coincidence_1, qdc_coincidence_2, edgecolor='k' )
ax.set_xlabel('QDC ch. %d'%ch_1)
ax.set_ylabel('QDC ch. %d'%ch_2)
plt.grid(alpha=0.1)
plt.plot([0,0],[600,600])
```
