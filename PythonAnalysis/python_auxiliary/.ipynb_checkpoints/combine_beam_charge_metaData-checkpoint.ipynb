## load run-by-run beam charges and combine them

### correct beam charge

there was an issue with the beam charge and Caleb for sure didnt took this into account for the 10.6 GeV plots. It was not corrected for pass1. This is the information from Silvia and also on the RGB website. It also depends which charge you are using (gated or ungated):

The workaround is "if run <6400   fc = 9.96025*fc".
More infos at:
https://clasweb.jlab.org/wiki/index.php/Run_Group_B#tab=Software_and_Data
REC::Event.beamCharge is the gated FCup and can be used directly. but if you are using the FC charge in your analysis, make sure to account for this 

(something along the lines of "if run <6400 fc = 9.96025*fc").

Therefore, the numbers from Caleb need to be divided by 9.96 for 10.6 GeV.


here are information about qadb access via clas12root in one of Dereks slides at https://clasweb.jlab.org/wiki/images/1/1f/Clas12Root1.7.pdf

You can cross check what you get there with the RUN:config info.



```python
rungroup = 'rgb';
# rungroup = 'rga';
data_path = '/Users/erezcohen/Desktop/data/BAND/metaData/' +rungroup + '/';
filelist = os.listdir(data_path)
run_beam_charges = pd.DataFrame()
for filename in filelist:
    if 'csv' in filename:
        run_beam_charge = pd.read_csv(data_path + filename)
        if (run_beam_charge.empty): continue
        run_beam_charge['Ebeam'] = beam_energy_from_run(int(run_beam_charge['runnum']),rungroup)
        run_beam_charges = pd.concat([run_beam_charges,run_beam_charge])        

Q = np.array(run_beam_charges['beam_charge'])
idcs = np.where(run_beam_charges.runnum<6400)
for idx in idcs: Q[idx] = 9.96025 * Q[idx]
run_beam_charges['Beam Charge [nC]'] = Q

run_beam_charges[['runnum','Beam Charge [nC]','Ebeam']].to_csv('/Users/erezcohen/Desktop/data/BAND/metaData/' 
                        + rungroup + '_all_runs.csv')
run_beam_charges        
```

```python
ylim = (0,1.1*np.max(run_beam_charges['Beam Charge [nC]']*1e-6))
fig = plt.figure(figsize=(22,7),tight_layout=True)
for Ebeam,E_idx in zip([10.2,10.6,10.4],[1,2,3]):
    beam_charges = run_beam_charges[run_beam_charges.Ebeam==Ebeam]
    ax = fig.add_subplot(1,3,E_idx)
    plt.errorbar(x=beam_charges.runnum, 
                 y=beam_charges['Beam Charge [nC]']*1e-6, yerr=beam_charges['Beam Charge [nC]']*0.03*1e-6,
                 marker='o', markeredgecolor='k',linestyle='',capthick=2
                );
    set_axes(ax,'run number','Charge [mC]',title='%.1f GeV'%Ebeam,
             do_add_legend=False,ylim=ylim,do_add_grid=True);
```
